```python
class Value:
    def __init__(self, value, _children=(), _op=""):
        self.data = value
        self._prev = set(_children)
        self._op = _op
    
    def __repr__(self):
        return f"Value = {self.data}"
    
    def __add__(self, other):
        out = Value(self.data + other.data, (self, other), "+")
        return out
    
    def __mul__(self, other):
        out = Value(self.data * other.data, (self, other), "*")
        return out
```

```python
a = Value(3) 
b = Value(3.3) 
c = Value(2)
a*b+c
```
